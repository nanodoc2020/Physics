## Problem 2.2

Find the electric field (magnitude and direction) a distance $z$ above the midpoint between equal and opposite charges ($\pm q$) a distance $d$ apart.

```python
img_data = plt.imread('griffiths_EM_2_2_diagram.JPG')
_=plt.figure(figsize=(20,5))
_=plt.imshow(img_data)
```

The diagram shows the situation with a negative charge, $-q$, on the left. 

### Equations

$$\vec{E} = \vec{E}_{+} + \vec{E}_{-} \hspace{2.5cm}\vec{E}=\frac{kq}{r^{2}}\hat{r}$$

$$\vec{E}_{+x} = |E|\sin{(\theta)}\hat{x} \hspace{2.5cm} \vec{E}_{-x}=|E|\sin{(\phi)}\hat{x}$$

$$\vec{E}_{+y}=|E|\cos{(\theta)}\hat{y} \hspace{2.5cm} \vec{E}_{-y}=|E|\cos{(\phi)}\hat{y} $$


$$\begin{gather} r = \sqrt{z^{2} + \left(\frac{d}{2}\right)^{2} } \\\\[5pt]
= z\sqrt{1 + \left(\frac{d}{2z}\right)^{2} } \hspace{1cm}\text{or}\hspace{1cm} \frac{d}{2}\sqrt{1 + \left(\frac{2z}{d}\right)^{2} }
\end{gather}$$

From the diagram we can deduce the following:\
$$\sin{(\theta)} = \frac{-\frac{d}{2}}{r} \hspace{2.5cm} \sin{(\phi)} = \frac{-\frac{d}{2}}{r} \hspace{2.5cm} \cos{(\theta)} = \frac{z}{r} \hspace{2.5cm} \cos{(\phi)} = \frac{z}{r}$$

### Finding $\vec{E_{x}}$

Making the proper substitutions for the $\hat{x}$ components, \
$$\begin{align}
\vec{E}_{+x} &= \frac{kq}{r^{2}} \cdot \frac{-d}{2r} \\\\[5pt]
&= -\frac{kqd}{2r^{3}} \text{,}
\end{align}$$

$$\begin{align}
\vec{E}_{-x} &= \frac{kq}{r^{2}} \cdot \frac{-d}{2r} \\\\[5pt]
&= -\frac{kqd}{2r^{3}} \text{,}
\end{align}$$

we find the total, $E_{x} = E_{+x} + E_{-x}$. 

$$\begin{align}
E_{x} &= -\frac{kqd}{2r^{3}} - \frac{kqd}{2r^{3}} \\\\[5pt]
&= -\frac{kqd}{r^{3}}
\end{align}$$

The $x$ component of the electric field is then $$ \vec{E_{x}} = -\frac{kqd}{r^{3}}\hat{x}.  $$

### Finding $\vec{E}_{y}$

Making the substitutions for the $\hat{y}$ components,

$$\begin{align}
\vec{E}_{+y} &= \frac{kq}{r^{2}} \cdot \frac{z}{r} \\\\[5pt]
&= \frac{kqz}{2r^{3}} \text{,}
\end{align}$$

$$\begin{align}
\vec{E}_{-y} &= \frac{kq}{r^{2}} \cdot \frac{z}{r} \\\\[5pt]
&= -\frac{kqz}{2r^{3}} \text{,}
\end{align}$$

we find the total, $E_{y} = E_{+y} + E_{-y}$. 

$$\begin{align}
E_{y} &= \frac{kqz}{2r^{3}} - \frac{kqz}{2r^{3}} \\\\[5pt]
&= 0
\end{align}$$

The $y$ component of the electric field is then $$ \vec{E_{y}} = 0 \hat{y}.  $$

### Finding the Total Field $\vec{E}$

The final field is found by summing the $x$ and $y$ components.

$$\begin{align}
\vec{E} &= \vec{E}_{x} + \vec{E}_{y} \\\\[5pt]
&= -\frac{kqd}{r^{3}} \hat{x} + 0 \hat{y} \\\\[5pt]
&= -\frac{kqd}{r^{3}} \hat{x}
\end{align}$$

Making the substitution for $r$, we arrive at the final expression.

$$\begin{align}
\vec{E} &= -\frac{ kqd }{ \left( \frac{d}{2}\sqrt{1 + \left(\frac{2z}{d}\right)^{2} } \right)^{3} } \hat{x} \\\\[5pt]
\vec{E}&= - \frac{8kq}{d^2 \left( 1 + \left(\frac{2z}{d}\right)^{2} \right)^{\frac{3}{2}}} \hat{x}
\end{align}$$

We can now explore some values and graphs of the equation.

```python
# Set up the variables
z= .5
d = .1
k = 1 / (4 * constants.pi * constants.epsilon_0)
q1 = 3 * constants.elementary_charge
```

```python
#Define a function to calculate the field
def E_field(z=.01, d=.05, q=constants.elementary_charge):
    E = 8 * k * q / (d**2 * (1 + (2 * z / d) ** 2) ** (3 / 2))
    return E
```

```python
# Create a list of points to plot for changing d and z
Ed_list = []
Ez_list = []
for i in np.arange(.0001,.02,.0001):
    E_d = E_field(z=.001, d=i, q=q1)
    E_z = E_field(z=i, d=.003, q=q1)
    Ed_list.append(E_d)
    Ez_list.append(E_z)

```

```python
# Plot the points
_=plt.figure(figsize=(14,5))
_=plt.plot(Ed_list)
_=plt.plot(Ez_list)
_=plt.axvline(Ed_list.index(max(Ed_list)), color='green', linestyle='--', lw='1')
_=plt.title('Field of Two Opposite-Sign Point Charges')
_=plt.xlabel('x (.1mm)')
_=plt.ylabel('Field Strength (N/C)')
_=plt.legend(['d','z'])
```

## Conclusion

The final expression shows the expected reduction in field strength for increasing $z$, and a surprising curve for increasing charge separation $d$. This makes sense, because if the two charges could be superimposed upon one another, the electric field at point $P$ would completely cancel. Thus, as the identical but opposite charges are moved apart, the $x$ components begin to add together while the $y$ components completely cancel out at all times. The field reaches maximum strength at $d = 1.3mm$ for fixed height($z$) of $1mm$, after which the additive effects are overcome by the exponentiated terms in the denominator.
